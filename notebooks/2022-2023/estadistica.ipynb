```python
url = 'https://rfevb-web.dataproject.com/Statistics.aspx?ID=111&PID=137'
driver = webdriver.Chrome(options = opciones)
driver.get(url)
```

```python
driver.find_element(By.XPATH,'//*[@id="RTS_Statistics_StatsType"]/div/ul/li[2]/a/span/span').click()
```

```python
resultado= driver.find_element(By.XPATH,'//*[@id="Content_Main_RPL_Master"]/div[6]').text.split('\n')
```

```python
columnas = ['Equipo', 'Jugados', 'Sets_Jugados', 'Total_Puntos', 'Break_Puntos', 'Puntos_Ganados',
            'Puntos_Perdidos', 'Total_Saque', 'Puntos_Saque', 'Error_Saque', 'Puntos_por_set_Saque',
            'Efic_Saque', 'Total_Recepcion', 'Error_Recepcion', 'Negativo_Recepcion', 'Positivo_Recepcion',
            'Excelente_Recepcion', 'Efic_Recepcion', 'Total_Ataque', 'Error_Ataque', 'Ataque_Bloqueado',
            'Positivo_Ataque', 'Excelente_Ataque', 'Efic_Ataque', 'Toque_Red_Bloqueo', 'Puntos_de_Bloqueo',
            'Puntos_Set_Bloqueo']

trozos = [resultado[i:i+27] for i in range(2, len(resultado), 27)]
estadistica = pd.DataFrame(trozos, columns=columnas)
estadistica = estadistica.drop(index=12)
estadistica
```

```python
mapeo = {
    'Arenal Emevé':'Arenal Emevé',
    'Barça Voleibol':'Barça Voleibol',
    'Club Vóley Palma':'Voley Palma',
    'Conectabalear CV Manacor':'Manacor',
    'CV Guaguas':'Guaguas',
    'Léleman Conqueridor Valencia':'Conqueridor Valencia',
    'Melilla Sport Capital':'CV Melilla',
    'Pamesa Teruel Voleibol':'CV Teruel',
    'Rotogal Boiro':'Rotogal Boiro',
    'Rio Duero Soria':'Rio Duero Soria',
    'Unicaja Costa de Almeria':'Unicaja Almería',
    'Voley Textil Santanderina':'Voley Textil Santanderina'
}

estadistica['Equipo'] = estadistica['Equipo'].str.strip().map(mapeo)

estadistica
```

```python
#Extraccion de jugadores.
df= pd.read_csv('../../data/2022-2023/equipo.csv')
df
```

```python
estadistica = pd.merge(estadistica, df[['Equipo', 'id_equipo']], left_on='Equipo', right_on='Equipo', how='left')
estadistica.head()
```
