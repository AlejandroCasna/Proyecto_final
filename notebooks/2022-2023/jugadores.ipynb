```python
import selenium
import pandas as pd
import numpy as np
import time
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
opciones = Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('--incognito')
opciones.add_argument('--start-maximized')
```

```python
url = 'https://rfevb-web.dataproject.com/CompetitionHome.aspx?ID=111'
driver = webdriver.Chrome(options = opciones)
driver.get(url)
```

```python
driver.find_element(By.XPATH,'//*[@id="ctl00_RM_CompetitionArea_Main"]/ul/li[5]/a').click()
```

```python
superLiga_Masculino=[]
for i in range(12):
                
    e = driver.find_elements(By.CLASS_NAME, 'rlvI')[i]
                
    print(e.text)
                
    e.click()
                
    print('Estoy dentro.')
                
    time.sleep(1)
            
    superLiga_Masculino.append(driver.find_elements(By.CLASS_NAME , 't-row')[3].text.split('\n'))
                
    print('Extraigo')
                
    time.sleep(1)
                
    driver.get('https://rfevb-web.dataproject.com/CompetitionTeamSearch.aspx?ID=111')
```

```python
superLiga_Masculino[0][0:8]
```

```python
len(superLiga_Masculino)
```

```python
columnas = ['Dorsal', 'Nombre', '1', 'Posición', 'Altura', 'Año de nacimiento', 'Alcance en ataque', 'Alcance en bloqueo']
filas_organizadas = []

for equipo in superLiga_Masculino:
    datos_filas = equipo[7:]
    num_filas = len(datos_filas) // len(columnas)
    filas_equipo = [datos_filas[i:i + len(columnas)] for i in range(0, len(datos_filas), len(columnas))]
    filas_organizadas.extend(filas_equipo)

jugadores = pd.DataFrame(filas_organizadas, columns=columnas)
jugadores.head()
```

```python
jugadores = jugadores.drop('1', axis=1)

jugadores
```

```python
jugadores['id_jugador'] = range(1, len(jugadores) + 1)
jugadores
```

```python
jugadores.to_csv('../../data/2022-2023/jugadores.csv', index=False)
```
