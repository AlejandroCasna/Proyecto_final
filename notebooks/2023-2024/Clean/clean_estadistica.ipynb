```python
columnas = ['Equipo', 'Jugados', 'Sets_Jugados', 'Total_Puntos', 'Break_Puntos', 'Puntos_Ganados',
            'Puntos_Perdidos', 'Total_Saque', 'Puntos_Saque', 'Error_Saque', 'Puntos_por_set_Saque',
            'Efic_Saque', 'Total_Recepcion', 'Error_Recepcion', 'Negativo_Recepcion', 'Positivo_Recepcion',
            'Excelente_Recepcion', 'Efic_Recepcion', 'Total_Ataque', 'Error_Ataque', 'Ataque_Bloqueado',
            'Positivo_Ataque', 'Excelente_Ataque', 'Efic_Ataque', 'Toque_Red_Bloqueo', 'Puntos_de_Bloqueo',
            'Puntos_Set_Bloqueo']

trozos = [resultado[i:i+27] for i in range(2, len(resultado), 27)]
estadistica = pd.DataFrame(trozos, columns=columnas)
estadistica
```

```python
#Extraccion de jugadores.
df= pd.read_csv('../../../data/2023-2024/Equipos.csv')
df
```

```python
mapeo_equipo = {
    'Grupo Herce Soria': 'Rio Duero Soria',
    'Club Vóley Palma': 'Voley Palma',
    'Conectabalear CV Manacor': 'Manacor',
    'CV Guaguas': 'Guaguas',
    'CV San Roque': 'San Roque - Batán',
    'Pamesa Teruel Voleibol': 'CV Teruel',
    'Unicaja Costa de Almería': 'Unicaja Almería',
    'UPV Leleman Conqueridor': 'Conqueridor Valencia',
    'Volei Villena Petrer': 'Santo Domingo VB Petrer'
}

estadistica['Equipo'] = estadistica['Equipo'].map(mapeo_equipo).fillna(estadistica['Equipo'])
```

```python
result = pd.merge(estadistica, df[['Nombre', 'id_equipo']], left_on='Equipo', right_on='Nombre', how='left')
result.head()
```

```python
result = result.drop(columns=['Nombre'])
```

```python
columns_to_convert = ['Efic_Ataque', 'Excelente_Ataque', 'Efic_Recepcion', 'Excelente_Recepcion', 'Puntos_por_set_Saque', 'Efic_Saque','Puntos_Set_Bloqueo']
result[columns_to_convert] = result[columns_to_convert].replace({'%': '', ',': '.'}, regex=True).astype(float)
result.info()
```

```python
convertir = result.columns.difference(['Nombre'])
columns_to_convert = ['Efic_Ataque', 'Excelente_Ataque', 'Efic_Recepcion', 'Excelente_Recepcion']
result[columns_to_convert] = result[columns_to_convert].replace('%', '', regex=True)
result[convertir] = result[convertir].applymap(lambda x: int(x) if str(x).isdigit() else x)

result.info()
```
