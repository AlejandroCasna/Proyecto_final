```python
url = 'https://rfevb-web.dataproject.com/CompetitionTeamSearch.aspx?ID=124'
parametro = 'jugadores'
datos= scraping(url,parametro)
```

```python
dataframes = []
for lista in datos:
    columnas = lista[:7]
    datos_filas = lista[7:]
    num_filas = len(datos_filas) // len(columnas)
    filas_organizadas = [datos_filas[i:i+len(columnas)] for i in range(0, len(datos_filas), len(columnas))]
    df = pd.DataFrame(filas_organizadas, columns=columnas)
    dataframes.append(df)
jugadores = pd.concat(dataframes, ignore_index=True)
jugadores
```

```python
jugadores['id_jugador'] = [str(i) + '-2023-2024' for i in range(1, len(jugadores) + 1)]
jugadores['temporada'] = '2023-2024'
jugadores.columns = jugadores.columns.str.replace(' ', '_')
jugadores.loc[0:11, 'id_equipo'] = '1-2023-2024'
jugadores.loc[12:26, 'id_equipo'] = '2-2023-2024'
jugadores.loc[27:39, 'id_equipo'] = '3-2023-2024'
jugadores.loc[40:52, 'id_equipo'] = '4-2023-2024'
jugadores.loc[53:66, 'id_equipo'] = '5-2023-2024'
jugadores.loc[67:87, 'id_equipo'] = '6-2023-2024'
jugadores.loc[88:93, 'id_equipo'] = '7-2023-2024'
jugadores.loc[94:107, 'id_equipo'] = '8-2023-2024'
jugadores.loc[108:125, 'id_equipo'] = '9-2023-2024'
jugadores.loc[126:138, 'id_equipo'] = '10-2023-2024'
jugadores.loc[139:152, 'id_equipo'] = '11-2023-2024'
jugadores.loc[153:, 'id_equipo'] = '12-2023-2024'
jugadores.replace('-', '0', inplace=True)
jugadores[['Alcance_en_ataque', 'Alcance_en_bloqueo', 'Altura']] = jugadores[['Alcance_en_ataque', 'Alcance_en_bloqueo', 'Altura']].astype(float)
jugadores['Año_de_nacimiento'] = pd.to_numeric(jugadores['Año_de_nacimiento'], errors='coerce').fillna(0).astype(int)
jugadores.rename(columns={'#': 'Dorsal'}, inplace=True)
jugadores.head()
```
