# definimos DataFrame jugadores completo


```python
Jugadores = pd.DataFrame()

# Iterar a través de las 12 sub-listas
for sublist in superLiga_Masculino:
    # Definir las columnas y los datos
    columns = sublist[0:4]
    data = sublist[4:]
    
    # Dividir los datos en lotes de 4 (ya que cada jugador tiene 4 atributos)
    player_data = [data[i:i+4] for i in range(0, len(data), 4)]
    
    # Crear un DataFrame temporal para esta sub-lista
    temp_df = pd.DataFrame(player_data, columns=columns)
    
    # Cambiar el nombre de la columna '#'
    temp_df = temp_df.rename(columns={'#': 'Dorsal'})
    
    # Concatenar el DataFrame temporal con el DataFrame principal
    Jugadores = pd.concat([Jugadores, temp_df], ignore_index=True)

# Mostrar el DataFrame final
Jugadores
```

```python
Jugadores['ID'] = Jugadores.index + 1
Jugadores
```
