```python
import pandas as pd
import streamlit as st
import plotly.express as px

# Cargar los dataframes
df_estadistica = pd.read_csv('posgresSQL\data\estadistica.csv')
columnas_seleccionadas = ['Equipo', 'Efic_Saque', 'Efic_Recepcion', 'Efic_Ataque', 'Puntos_Set_Bloqueo', 'temporada']
df_estadistica = df_estadistica[columnas_seleccionadas]

# Función para el gráfico de araña
def araña(equipo):
    # Filtrar las filas donde 'Equipo' sea igual al equipo seleccionado
    df_selected_team = df_estadistica[df_estadistica['Equipo'] == equipo]

    # Aplicar melt al DataFrame filtrado
    df_melt = pd.melt(df_selected_team, id_vars=['Equipo'], var_name='theta', value_name='r')

    # Crear el gráfico de araña
    fig = px.line_polar(df_melt, r='r', theta='theta', line_close=True, color='Equipo', line_group='Equipo')

    # Actualizar las trazas para rellenar el área bajo la línea
    fig.update_traces(fill='toself')

    fig.update_layout(
        paper_bgcolor='rgba(0,0,0,0)',  # Fondo transparente
        plot_bgcolor='rgba(0,0,0,0)',  # Fondo del gráfico transparente
        font=dict(size=20, color='black'),  # Tamaño de fuente más grande para los números en los ejes
    )
    fig.update_polars(
        angularaxis=dict(
            tickfont=dict(size=25, color='black'),  # Tamaño de fuente más grande para los títulos de los ejes
            linecolor='black',  # Color negro para las líneas del eje
        ),
        radialaxis=dict(
            tickfont=dict(size=20, color='black'),  # Tamaño de fuente más grande para los números en los ejes
            linecolor='black',  # Color negro para las líneas del eje
        )
    )
    st.plotly_chart(fig)

# Función para el gráfico de dona
def dona():
    # Datos de ejemplo
    data = {'Categoría': ['A', 'B', 'C'],
            'Valores': [30, 40, 30]}

    df = pd.DataFrame(data)

    # Gráfico de dona con Plotly Express
    fig = px.pie(df, names='Categoría', values='Valores', hole=0.3)

    # Mostrar el gráfico en Streamlit
    st.plotly_chart(fig)

# Función para el gráfico de pie
def pie_chart():
    # Datos de ejemplo
    data = {'Categoría': ['A', 'B', 'C'],
            'Valores': [30, 40, 30]}

    df = pd.DataFrame(data)

    # Gráfico de pie con Plotly Express
    fig = px.pie(df, names='Categoría', values='Valores')

    # Mostrar el gráfico en Streamlit
    st.plotly_chart(fig)

# Ejemplo de uso
araña('Equipo1')  # Reemplaza 'Equipo1' con el equipo que desees
dona()
pie_chart()
```
