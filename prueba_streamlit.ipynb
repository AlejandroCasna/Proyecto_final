```python
import pandas as pd
import streamlit as st
import plotly.express as px

# Cargar los dataframes
df_estadistica = pd.read_csv('posgresSQL\data\estadistica.csv')
columnas_seleccionadas = ['Equipo', 'Efic_Saque', 'Efic_Recepcion', 'Efic_Ataque', 'Puntos_Set_Bloqueo', 'temporada']
df_estadistica = df_estadistica[columnas_seleccionadas]

# Funci칩n para el gr치fico de ara침a
def ara침a(equipo):
    # Filtrar las filas donde 'Equipo' sea igual al equipo seleccionado
    df_selected_team = df_estadistica[df_estadistica['Equipo'] == equipo]

    # Aplicar melt al DataFrame filtrado
    df_melt = pd.melt(df_selected_team, id_vars=['Equipo'], var_name='theta', value_name='r')

    # Crear el gr치fico de ara침a
    fig = px.line_polar(df_melt, r='r', theta='theta', line_close=True, color='Equipo', line_group='Equipo')

    # Actualizar las trazas para rellenar el 치rea bajo la l칤nea
    fig.update_traces(fill='toself')

    fig.update_layout(
        paper_bgcolor='rgba(0,0,0,0)',  # Fondo transparente
        plot_bgcolor='rgba(0,0,0,0)',  # Fondo del gr치fico transparente
        font=dict(size=20, color='black'),  # Tama침o de fuente m치s grande para los n칰meros en los ejes
    )
    fig.update_polars(
        angularaxis=dict(
            tickfont=dict(size=25, color='black'),  # Tama침o de fuente m치s grande para los t칤tulos de los ejes
            linecolor='black',  # Color negro para las l칤neas del eje
        ),
        radialaxis=dict(
            tickfont=dict(size=20, color='black'),  # Tama침o de fuente m치s grande para los n칰meros en los ejes
            linecolor='black',  # Color negro para las l칤neas del eje
        )
    )
    st.plotly_chart(fig)

# Funci칩n para el gr치fico de dona
def dona():
    # Datos de ejemplo
    data = {'Categor칤a': ['A', 'B', 'C'],
            'Valores': [30, 40, 30]}

    df = pd.DataFrame(data)

    # Gr치fico de dona con Plotly Express
    fig = px.pie(df, names='Categor칤a', values='Valores', hole=0.3)

    # Mostrar el gr치fico en Streamlit
    st.plotly_chart(fig)

# Funci칩n para el gr치fico de pie
def pie_chart():
    # Datos de ejemplo
    data = {'Categor칤a': ['A', 'B', 'C'],
            'Valores': [30, 40, 30]}

    df = pd.DataFrame(data)

    # Gr치fico de pie con Plotly Express
    fig = px.pie(df, names='Categor칤a', values='Valores')

    # Mostrar el gr치fico en Streamlit
    st.plotly_chart(fig)

# Ejemplo de uso
aran팪a('Equipo1')  # Reemplaza 'Equipo1' con el equipo que desees
dona()
pie_chart()
```
